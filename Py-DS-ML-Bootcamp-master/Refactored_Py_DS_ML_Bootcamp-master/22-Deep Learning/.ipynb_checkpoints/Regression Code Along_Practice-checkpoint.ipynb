```python
#we can categorize expensive zipcodes and not expensive ones and so on..
df = df.drop("zipcode", axis=1)
```

```python
df["yr_renovated"].value_counts()
```

```python
df["sqft_basement"].value_counts()
```

```python
# returns numpy arrays with .values, which is more secure 
X = df.drop("price",axis=1).values
y = df["price"].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
```

```python
X_train.shape
```

```python
model = Sequential()

model.add(Dense(19,activation="relu"))
model.add(Dense(19,activation="relu"))
model.add(Dense(19,activation="relu"))
model.add(Dense(19,activation="relu"))

model.add(Dense(1))

model.compile(optimizer="adam",loss="mse")
```

```python
#check if we did .values to pd data before
#the smaller the batch size, the longer it will take, but less likely the overfitting occures
model.fit(x=X_train,y=y_train,validation_data=(X_test,y_test),
         batch_size=128,epochs=400)
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
#if orange line starts to spike up, it is overfitting, but now we can further increase the number of epoch even more
losses.plot()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, explained_variance_score
```

```python
predictions = model.predict(X_test)
```

```python
np.sqrt(mean_squared_error(y_test,predictions))
```

```python
mean_absolute_error(y_test,predictions)
```

```python
df["price"].describe()
```

```python
#off by around 20%, not so great
5.402966e+05
```

```python
explained_variance_score(y_test,predictions)
```
